```python
import json
import os
import pandas as pd

# file_path = os.path.join("conceptual_model", "aiod_model.json")
file_path = os.path.join("conceptual_model", "ontology_data_with_inheritance.json")

with open(file_path, "r") as f:
    conceptual_model = json.load(f)
```

```python
missing_entities_df = pd.read_csv("comparision_results_noinheritance/missing_entities.csv")
extra_entities_df = pd.read_csv("comparision_results_noinheritance/extra_entities.csv")
missing_properties_df = pd.read_csv("comparision_results_noinheritance/missing_properties.csv")
extra_properties_df = pd.read_csv("comparision_results_noinheritance/extra_properties.csv")
```

```python
entity_name = "Project"

index_miss = missing_properties_df[missing_properties_df["Entity"] == entity_name.lower()].index
index_extra = extra_properties_df[extra_properties_df["Entity"] == entity_name.lower()].index


df_combined = pd.concat(
    [missing_properties_df.iloc[index_miss[0]], extra_properties_df.loc[index_extra[0]]], axis=1
)
df_combined.columns = ["Missing", "Extra"]
# Sort each column alphabetically
df_combined = df_combined.apply(lambda col: col.sort_values())
df_combined.dropna(how="all")
```
